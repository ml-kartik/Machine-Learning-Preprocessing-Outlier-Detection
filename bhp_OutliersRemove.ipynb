```python
import pandas as pd
df = pd.read_csv("bhp.csv")
df.head()
```

```python
df.shape
```

# Step 1 : Remove Extreme Outliers Using Percentile Method

```python
lower_bound , upper_bound = df.price_per_sqft.quantile([0.001,0.999])
df[df.price_per_sqft > upper_bound] 
```

```python
df[df.price_per_sqft < lower_bound]
```

```python
df1 = df[(df.price_per_sqft < upper_bound) &  (df.price_per_sqft > lower_bound)]
```

```python
df1.shape
```

# Remove Outliers Using 4 standard deviation from new Data Frame (df1)

```python
df1.describe()
```

```python
upper_limit = df1.price_per_sqft.mean() + 4* df1.price_per_sqft.std()
lower_limit = df1.price_per_sqft.mean() - 4* df1.price_per_sqft.std()
df1[df1.price_per_sqft > upper_limit]
```

```python
df1[df1.price_per_sqft < lower_limit]
```

```python
df1[(df1.price_per_sqft > upper_limit) | (df1.price_per_sqft < lower_limit)]
```

```python
df2 = df1[(df1.price_per_sqft < upper_limit) & (df1.price_per_sqft > lower_limit)]
df2
```

```python
df2.describe()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
```

```python
plt.hist(df2.price_per_sqft, bins = 20, rwidth = 0.8, density = True )
plt.xlabel("Price per Sqft")
plt.ylabel('Count')


rng = np.arange(df2.price_per_sqft.min() , df2.price_per_sqft.max() , 0.1 )
plt.plot(rng, norm.pdf(rng,df2.price_per_sqft.mean(),df2.price_per_sqft.std()))
```

# By using Z score Method

```python
df1.describe()
```

```python
df1['zscore']  = (df1.price_per_sqft - df1.price_per_sqft.mean()) / df1.price_per_sqft.std()
```

```python
df1
```

```python
df1[df1.zscore > 4]
```

```python
df1[df1.zscore <- 4]
```

```python
df_no_outliers = df1[(df1.zscore < 4) & (df1.zscore > -4)]
df_no_outliers
```
