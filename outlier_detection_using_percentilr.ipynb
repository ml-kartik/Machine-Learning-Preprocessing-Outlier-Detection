```python
import pandas as pd
df = pd.read_csv("name_height_outlier.csv")
df
```

```python
max_threshold = df["Height_cm"].quantile(0.95) #finds the 95th percentile value.
max_threshold
```

```python
df[df['Height_cm']>max_threshold]
```

```python
min_threshold = df["Height_cm"].quantile(0.05) #finds the 5th percentile value.
min_threshold
```

```python
df[df['Height_cm']<min_threshold]
```

```python
df[(df["Height_cm"]<max_threshold) & (df["Height_cm"]>min_threshold)]
```

# Lets do it on Complex Data

```python
df = pd.read_csv("Housing.csv")
df
```

```python
df.shape
```

```python
df.describe()
```

```python
df["price_per_sqft"] = df["price"] / df["area"]
df
```

```python
df.describe()
```

```python
min_threshold, max_threshold = df.price_per_sqft.quantile([0.001,0.999])
min_threshold, max_threshold
```

```python
df[df.price_per_sqft < min_threshold]
```

```python
df[df.price_per_sqft > max_threshold]
```

```python
df2 = df[(df.price_per_sqft > min_threshold) & (df.price_per_sqft < max_threshold)]
df2
```
