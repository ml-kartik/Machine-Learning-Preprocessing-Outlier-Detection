# We want the height to be from 4.5 ft to 6ft

```python
import pandas as pd
df = pd.read_csv('weight-height-outliers.csv')
df['height'] = df["Height"]/12
df.drop(['Weight','Height'],axis =1, inplace = True)
df.head()
```

```python
df.shape
```

```python
df.describe()
```

# Removing outliers using IQR

```python
Q1 = df.height.quantile(0.25)
Q3 = df.height.quantile(0.75)
```

```python
IQR = Q3-Q1
IQR
```

```python
lower_bound = Q1 - 1.5*IQR
lower_bound
```

```python
upper_bound = Q1 + 1.5*IQR
upper_bound
```

```python
df1 = df[(df.height < upper_bound) & (df.height > lower_bound)]
df1.shape
```

```python
df.shape[0] - df1.shape[0]
```
